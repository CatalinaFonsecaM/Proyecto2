```python

data_health = wb.data.DataFrame('SH.XPD.CHEX.PC.CD',
                                  economy=['ARG','BOL','BRA','CAN','CHL','COL','CRI','ECU','MEX','PRY','PER','USA','URY','VEN'],
                                  time=range(2002,2019),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
data_health
```

```python
data_gdp = wb.data.DataFrame('NY.GDP.PCAP.CD',
                                  economy=['ARG','BOL','BRA','CAN','CHL','COL','CRI','ECU','MEX','PRY','PER','USA','URY','VEN'],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
data_gdp
```

```python
data_water = wb.data.DataFrame('ER.H2O.INTR.PC',
                                  economy=['ARG','BOL','BRA','CAN','CHL','COL','CRI','ECU','MEX','PRY','PER','USA','URY','VEN'],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
data_water
```

```python
data_nutrition = wb.data.DataFrame('SN.ITK.DEFC.ZS',
                                  economy=['ARG','BOL','BRA','CAN','CHL','COL','CRI','ECU','MEX','PRY','PER','USA','URY','VEN'],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
data_nutrition
```

```python
data_unemployment = wb.data.DataFrame('SL.UEM.TOTL.ZS',
                                  economy=['ARG','BOL','BRA','CAN','CHL','COL','CRI','ECU','MEX','PRY','PER','USA','URY','VEN'],
                                  time=range(2002,2020),
                                  labels=True).reset_index(drop=True).set_index('Country').unstack().reset_index()
data_unemployment
```

```python
df=pd.merge(data_unemployment,data_gdp,how='outer',on=['level_0','Country']).merge(data_health,how='outer',on=['level_0','Country']).merge(data_water,how='outer',on=['level_0','Country']).merge(data_nutrition,how='outer',on=['level_0','Country'])
```

```python
def function2(row):
  ingreso = row['pib']
  if ingreso <= 8000:
    return 'ingreso_bajo'
  elif ingreso <= 16000:
    return 'ingreso_medio'
  else:
    return 'ingreso_alto'
df['nivel_ingreso'] = df.apply(function2, axis=1)
df['nivel_ingreso'].describe()
```
